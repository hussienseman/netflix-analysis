```python
import pandas as pd

url = "https://raw.githubusercontent.com/hussienseman/netflix-analysis/refs/heads/main/netflix_titles.csv"
df = pd.read_csv(url)

df.head()
```

```python
# See basic info about the dataset
df.info()

# How many rows and columns?
df.shape

# Check for missing values
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar', color=['red','blue'])
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```
